### Variables Description

**Category** groups apps into 33 types (Family, Game, Tools, Business, etc.). **Rating** is the average star score out of 5. **Reviews** is the total number of written reviews. **Size** (cleaned into **Size_MB**) is the install file size in megabytes — some apps report "Varies with device" since size depends on the user's phone, so this is treated as missing. **Installs** is a bucketed count ("10,000+", "500,000+", etc.) which we convert to a plain number for analysis. **Type** is Free or Paid, and **Price** is the dollar cost (0 for free apps). **Content Rating** is the target age group, and **Genres** is a finer sub-category than **Category**. On the reviews side, **Sentiment** is a Positive/Negative/Neutral label, **Sentiment_Polarity** ranges from -1 (very negative) to +1 (very positive), and **Sentiment_Subjectivity** ranges from 0 (purely factual) to 1 (purely opinion).

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Step 1: Drop the single corrupted row.
# 'Life Made WI-Fi Touchscreen Photo Frame' has Category == '1.9' and Rating == 19,
# which is impossible (ratings cap at 5). A missing Category value shifted every
# later column left by one position, corrupting the whole row.
play = play[play['Category'] != '1.9'].copy()

# Step 2: Reviews -> integer
play['Reviews'] = pd.to_numeric(play['Reviews'], errors='coerce').astype('Int64')

# Step 3: Installs -> integer  (e.g. "10,000+" -> 10000)
play['Installs'] = (
    play['Installs'].astype(str)
    .str.replace('+', '', regex=False)
    .str.replace(',', '', regex=False)
)
play['Installs'] = pd.to_numeric(play['Installs'], errors='coerce')

# Step 4: Price -> float  (e.g. "$4.99" -> 4.99)
play['Price'] = play['Price'].astype(str).str.replace('$', '', regex=False)
play['Price'] = pd.to_numeric(play['Price'], errors='coerce')

# Step 5: Size -> Size_MB (float). "Varies with device" becomes NaN.
def size_to_mb(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip()
    if x == 'Varies with device':
        return np.nan
    if x.endswith('M'):
        return float(x[:-1])
    if x.endswith('k'):
        return float(x[:-1]) / 1024
    try:
        return float(x)
    except ValueError:
        return np.nan

play['Size_MB'] = play['Size'].apply(size_to_mb)

# Step 6: Fix the Type column (one stray '0' value, one blank value)
play.loc[play['Type'] == '0', 'Type'] = 'Free'
fill_values = play['Price'].apply(lambda p: 'Paid' if p > 0 else 'Free')
play['Type'] = play['Type'].fillna(fill_values)

# Step 7: Last Updated -> datetime, extract year
play['Last Updated'] = pd.to_datetime(play['Last Updated'], errors='coerce')
play['Last_Updated_Year'] = play['Last Updated'].dt.year

# Step 8: Drop duplicate apps, keep the first occurrence
play = play.drop_duplicates(subset=['App'], keep='first').copy()

print("Final Play Store shape:", play.shape)

# Step 9: Clean the User Reviews dataset
# Drop blank review rows (no text/sentiment) and exact duplicate rows
rev = rev.dropna(subset=['Translated_Review', 'Sentiment']).copy()
rev = rev.drop_duplicates().copy()
print("Final User Reviews shape:", rev.shape)
````

### What all manipulations have you done and insights you found?

The dataset needed quite a bit of cleanup before it was analysis-ready:
- Removed 1 corrupted row where a shifted column produced an impossible Rating of 19.
- Converted `Reviews`, `Installs`, `Price`, and `Size` from text into clean numeric columns.
- Fixed a stray `'0'` value and 1 missing value in the `Type` column using `Price` as ground truth.
- Parsed `Last Updated` into a real date and extracted the year for time-trend analysis.
- Removed 1,181 duplicate app rows (same app scraped more than once).
- In the reviews dataset, removed 26,868 blank rows and 7,735 duplicate rows.

After cleaning: **9,659** unique apps remain in the Play Store dataset, and **29,692** usable reviews remain in the User Reviews dataset, covering **816** apps that exist in both datasets at once (used later for the sentiment-vs-rating comparison).
